```python
import urllib2
from bs4 import BeautifulSoup
import csv
import unicodedata
import pandas as pd
import IPython
import numpy as np
import scipy as sp
```

```python
df = pd.DataFrame(columns= ['First name','Last name','Grad School','Faculty School','Department'])
index = ['First name','Last name','Grad School','Faculty School','Department']
```

```python
directorypage = urllib2.urlopen('http://jindal.utdallas.edu/faculty/').read()
directorysoup = BeautifulSoup(directorypage)
i = 0
for inp in directorysoup.select(".allPeople")[0].find_all('input'):
    lastname = inp.find_parent('li').find('p').text.split(',')[0]
    firstname = inp.find_parent('li').find('p').text.split(',')[1]
    facultyURL = 'http://jindal.utdallas.edu/faculty/'+inp['value']
    facultypage = urllib2.urlopen(facultyURL).read()
    facultysoup = BeautifulSoup(facultypage)
    try:
        education = facultysoup.find_all("div",text="Education")[0].find_next_sibling("div").text.split(',')[1].encode('ascii')
    except:
        print "Error extracting: ",firstname,lastname
        continue
    row = []
    row.append(firstname.encode('ascii','ignore'))
    row.append(lastname.encode('ascii','ignore'))
    row.append(education.encode('ascii','ignore'))
    row.append('University of Texas, Dallas')
    row.append('Management')
    facultydf = pd.DataFrame([row], columns=index)
    df = df.append(facultydf, ignore_index=True) 
    i+=1
    if (i==200):
        break
print i
print row
df.to_csv("garimella_scraping.csv")
df
```
