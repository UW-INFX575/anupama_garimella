```python
import nltk #Use the nltk package to perform NLP tasks
import csv
import itertools
```

```python
from nltk.corpus import stopwords
from nltk.stem import *
from nltk.stem.porter import *
from nltk.util import ngrams
from itertools import groupby
```

```python
#Create empty lists to store n-grams from all 10 files combined
total_unigrams = []
total_bigrams = []
total_trigrams = []

for i in range(0,10): #Iterate over each file
    filename = str(6334220+i) #Generate the file name dynamically for each iteration
    
    #Open and read file
    f = open("C:\\Users\\Anupama\\OneDrive\\INFX575\\Assignment2\\textfiles\\"+filename+".txt", "r")
    textfile = f.read()

    #Remove stop words
    stop = stopwords.words('english')
    nonstop_words = [i for i in textfile.split() if i not in stop] #Store all words that are not stop words

    #Stem words using the Porter Stemming algorithm
    stemmer = PorterStemmer()
    stemmed_words = [stemmer.stem(i).encode("utf8") for i in nonstop_words] #Store all stemmed words

    #Find n-grams
    unigrams = list(ngrams(stemmed_words, 1))
    bigrams = list(ngrams(stemmed_words, 2))
    trigrams = list(ngrams(stemmed_words, 3))    
    
    #Append the n-grams found to the combined list as well
    total_unigrams.append(unigrams)
    total_bigrams.append(bigrams)
    total_trigrams.append(trigrams)

    #Find n-gram frequencies and store them in dictionaries
    freq_unigrams = {x:unigrams.count(x) for x in unigrams}
    freq_bigrams = {x:bigrams.count(x) for x in bigrams}
    freq_trigrams = {x:trigrams.count(x) for x in trigrams}

    #Write unigram frequencies to csv
    ngramfilename = filename+"_unigrams.csv"
    writer = csv.writer(open(ngramfilename, 'wb'))
    for key, value in freq_unigrams.items():
       writer.writerow([key[0], value])

    #Write bigram frequencies to csv
    ngramfilename = filename+"_bigrams.csv"
    writer = csv.writer(open(ngramfilename, 'wb'))
    for key, value in freq_bigrams.items():
       writer.writerow([" ".join(map(str,key)), value])
        
    #Write trigram frequencies to csv
    with open(filename+"_trigrams.csv", 'wb') as f:
        writer = csv.writer(f)
        for key, value in freq_trigrams.items():
            writer.writerow([" ".join(map(str,key)), value])
            
#For the combined n-grams, collapse the list of lists into flat lists
merged_unigrams = list(itertools.chain(*total_unigrams))
merged_bigrams = list(itertools.chain(*total_bigrams))
merged_trigrams = list(itertools.chain(*total_trigrams))

#Find the combined frequencies of the n-grams
total_freq_unigrams = {x:merged_unigrams.count(x) for x in merged_unigrams}
total_freq_bigrams = {x:merged_bigrams.count(x) for x in merged_bigrams}
total_freq_trigrams = {x:merged_trigrams.count(x) for x in merged_trigrams}
    
#Write combined unigram frequencies to a file
with open("combined_unigrams.csv", 'wb') as f:
        writer = csv.writer(f)
        for key, value in total_freq_unigrams.items():
            writer.writerow([" ".join(map(str,key)), value])  
            
#Write combined bigram frequencies to a file
with open("combined_bigrams.csv", 'wb') as f:
        writer = csv.writer(f)
        for key, value in total_freq_bigrams.items():
            writer.writerow([" ".join(map(str,key)), value])            

#Write combined trigram frequencies to a file
with open("combined_trigrams.csv", 'wb') as f:
        writer = csv.writer(f)
        for key, value in total_freq_trigrams.items():
            writer.writerow([" ".join(map(str,key)), value])            


```
